```python
#!/usr/bin/env python 
# -*- coding: utf-8 -*-

# This program is being made for the project work 
# named as TELEGRAM BOT USING PYTHON :
# performed in the CYBER LAB
# of Central University of Punjab, Bathinda, 151001
# by UMAR FAROOQ | 18mtcysc17
# under the supervison of Er. SURINDER SINGH KHURANA Sir.

"""
Basic example for a bot that uses inline and outline keyboards (buttons).
along with the use of commands
"""

"""
This is based on API
API is the acronym for Application Programming Interface, 
which is a software intermediary that allows two applications to talk to each other.
"""

#NOTE: This bot works only on predefined commands

#importing dependencies
#-----------------------------------------------------------------
import logging
# if an error occurs, then it provides more insights than a stack trace by telling you what the state
# of the program was before it arrived at the line of code where the error occured.

import telegram
#It is used to import essentials that are required in accessing telegram 

#-----------------------------------------------------------------

#importing particular modules
#_________________________________________________________________
from telegram import InlineKeyboardButton, InlineKeyboardMarkup
#Here the Inline Keyboard Buttons are being imported from the telegram library.

from telegram.ext import Updater
#Updater is imported that is used to receive the updates from the Telegram .

from telegram.ext import CommandHandler
#Commandhandler is imported that is used to handle the commands.

from telegram.ext import CallbackQueryHandler
#Callbackqueryhandler is imported to handle the queries and replying withm with the answers.

#___________________________________________________________________


logging.basicConfig(format='%(asctime)s - %(name)s - %(levelname)s - %(message)s', level=logging.INFO)

#format: This is the format of the log message.

#level: The root logger will be set to the specified severity level.
#here info level is used

logger = logging.getLogger(__name__)
   

#defining /start command 
#here inline keyboard with 3 inline buttons is made
def start(update, context):

    keyboard = [[InlineKeyboardButton("About", callback_data='About: \n\n Hello there,\n\n I am bot and i work on few commands.'),
                                    #About button
                                    #it replies the user with the message that conatins a little introduction about this bot.
                 InlineKeyboardButton("Contact", callback_data='Contact: \n\n UMAR FAROOQ \n\n soulatf3@gmail.com \n\n 916005728432')],
                                    #Contact button
                                    #it replies with the contact details of myself
                [InlineKeyboardButton("CommandsB", callback_data='CommandsB:\n\n /start\n\n /search\n\n /cupb\n\n /sharec\n\n /sharel'),
                 
                                    #Commands button
                                    #these commands provide us with the buttons
                                    #it replies to the commnads that only are used in this bot 
                                    #/start 
                                    #/search 
                                    #/cupb 
                                    #/sharec 
                                    #/sharel
                 InlineKeyboardButton("CommandsT", callback_data='CommnadsT: \n\n /help\n\n /tele\n\n /about')]]

                                    #these commands provide us with the text only
                                    #/help
                                    #/about 
                                    #/tele
                    
    reply_markup = InlineKeyboardMarkup(keyboard)



    update.message.reply_text('Please choose:', reply_markup=reply_markup)  
    #it will send the send the user the heading of the keyboard that is "Please choose" 
    


#defining /search command 
def search(update, context):
    keyboard = [[InlineKeyboardButton("Bold", callback_data='Bold: \n\n Type @bold <message> \n\n Example: @bold cyberlab')],
                [InlineKeyboardButton("Wiki", callback_data='Wiki: \n\n Type @wiki <query> \n\n Example: @wiki cupb'),
                InlineKeyboardButton("Youtube", callback_data='Youtube: \n\n Type @vid <video_name> \n\n Example: @vid cupb')],
               [InlineKeyboardButton("Gif", callback_data='Gif: \n\n Type @gif <gif_name> \n\n Example: @gif cat'),
                InlineKeyboardButton("Bing", callback_data='Bing: \n\n Type @bing <image_name> \n\n Example: @vid cat'),
                InlineKeyboardButton("Pic", callback_data='Pic: \n\n Type @pic <image_name> \n\n Example: @pic lion')],
               [InlineKeyboardButton("Imdb", callback_data='IMDB: \n\n Type @imdb <movie_name> \n\n Example: @imdb hacker')]] 
    reply_markup = InlineKeyboardMarkup(keyboard)

    update.message.reply_text('Please choose:', reply_markup=reply_markup)
    
#defining /cupb command 
#this command is particularly related to Central University of Punjab, bathinda (CUPB)
#this command lets the user to know about this university
#it will provide the buttons that are linked to the CUPB's website
#also it will send the emails and phone numbers of university officials

def cupb(update, context):
    keyboard = [[InlineKeyboardButton("Website", callback_data='Website:', url='www.cup.edu.in')],
                
                [InlineKeyboardButton("About", callback_data='About CUPB:',url='www.cup.edu.in/about_CUPB.php'),
                
                InlineKeyboardButton("VC", callback_data='VCs Desk:', url='www.cup.edu.in/vice_chancellor_message.php'),
                
                InlineKeyboardButton("VML", callback_data='Vision | Mission | Logo:', url='www.cup.edu.in/vision_mission_logo.php')],
               
                [InlineKeyboardButton("Academics", callback_data='Schools and departments:', url='www.cup.edu.in/schools_centres.php')],
                
                [InlineKeyboardButton("E-mails", callback_data='E-mail: \n\n cupvc09@gmail.com \n\n daa@cup.ac.in \n\n sa@cup.ac.in'), 
                
                 InlineKeyboardButton("Phone", callback_data='Phone: \n\n | 9464946974 | 9417387817 |  9812058109 | 9569889560')]]
    
    reply_markup = InlineKeyboardMarkup(keyboard)

    update.message.reply_text('Please choose:', reply_markup=reply_markup)

#defining /sharec command 
#sharec = share_contact
#this command lets the user to share his/her conatct 
#(particulary the phone number registered with  the telegram accessing this bot) 
#this uses keyboard that is not inline


def sharec(update, context):

    reply_markup = telegram.ReplyKeyboardMarkup([[telegram.KeyboardButton('Share Contact', request_contact=True)]])
        #this creates a keyboard that is not inline 
        #it appears at the position of our phone's keyboard
        #named as Share Contact
        
    update.message.reply_text('Please Share:', reply_markup=reply_markup)
        #it will send the user with the message "Please Share"

        
#defining /sharel command   
#sharec = share_location
#this command lets the user to share his/her acual location  
#this uses keyboard that is not inline

def sharel(update, context):
        #this creates a keyboard that is not inline 
        #it appears at the position of our phone's keyboard
        #named as Share Location
   
    reply_markup = telegram.ReplyKeyboardMarkup([[telegram.KeyboardButton('Share Location', request_location=True)]])

    update.message.reply_text('Please Share:', reply_markup=reply_markup)    
        #it will send the user with the message "Please Share"
    
#defining action for the button 
#it defines what the buttons will do after they are tapped

def button(update, context):
    query = update.callback_query

    query.edit_message_text(text="Selected option: {}".format(query.data))
        #the message sent will be
        # "Selected Option:" 
        #along with the data that is being written in the callback_query part
    
#defining /help command 
def help(update, context):
    update.message.reply_text("Use:/start to test this bot.")


def about(update, context):
    update.message.reply_text("This bot is using Python Programming at the back-end.")

def tele(update, context):
    update.message.reply_text('www.telegram.org')    

def error(update, context):
    """Log Errors caused by Updates."""
    logger.warning('Update "%s" caused error "%s"', update, context.error)
    
# here the warning level of logging is used
# logger.warning() is creating a LogRecord that holds all the information of the event 
# and passing it to all the Handlers that it has
    
    
#the main function is defined here 
#here we have to call all the commands that are used in this program

def main():
    # Create the Updater and pass it your bot's token.
    # Make sure to set use_context=True to use the new context based callbacks
    # Post version 12 this will no longer be necessary
    updater = Updater("970313498:AAErMddkccOwNelWSaaGnMnUK15DgwYwbNo", use_context=True)
    
   # Its purpose is to receive the updates from Telegram and to deliver them to said dispatcher.
   # It also runs in a separate thread, so the user can interact with the bot, for example on the command line.
   # The dispatcher supports handlers for different kinds of data: 
   #Updates from Telegram, basic text commands and even arbitrary types. 
   # The updater can be started as a polling service
    

    updater.dispatcher.add_handler(CallbackQueryHandler(button))
    
    updater.dispatcher.add_error_handler(error)
    #if error occurs it updates the context error
    
    #below is the list of commands that are used in this bot
    
    updater.dispatcher.add_handler(CommandHandler('start', start))
    
    updater.dispatcher.add_handler(CommandHandler('help', help))
    
    updater.dispatcher.add_handler(CommandHandler('tele', tele))
    
    updater.dispatcher.add_handler(CommandHandler('about', about))
    
    updater.dispatcher.add_handler(CommandHandler('search',search))
    
    updater.dispatcher.add_handler(CommandHandler('cupb',cupb))
    
    updater.dispatcher.add_handler(CommandHandler('sharec',sharec))
   
    updater.dispatcher.add_handler(CommandHandler('sharel',sharel))
    
    
    # Start the Bot
    updater.start_polling()

    # Run the bot until the user presses Ctrl-C or the stops the process,
   
    updater.idle()


if __name__ == '__main__':
    main()
    

    
    
#__________-----------UMAR FAROOQ----------_______
#__________-----------18mtcysc17-----------_______




```
